```python
import pandas as pd 
import numpy as np

CSV_file_source = "bridges.data.version1.csv"
```

```python
# Reading the data set
# Group by RIVER, PURPOSE and MATERIAL respectivly and use count agg function on IDENTIF column then rename it. 
data = pd.read_csv(CSV_file_source, index_col=False, header=None, names=['IDENTIF', 'RIVER', 'LOCATION', 'ERECTED', 'PURPOSE', 'LENGTH', 'LANESCLEAR', 'T-OR-D', 'MATERIAL', 'SPAN', 'REL-L', 'TYPE'])
data = data.groupby(['RIVER', 'PURPOSE', 'MATERIAL']).agg({'IDENTIF': "count"}).rename(columns={'IDENTIF': 'How Many?'})
data
```

```python
# Reading the data set
# Convert LENGTH column to numeric if failed return NaN
# Group by PURPOSE and MATERIAL respectivly and use mean agg function on LENGTH column then rename it. 
data = pd.read_csv(CSV_file_source, index_col=False, header=None, names=['IDENTIF', 'RIVER', 'LOCATION', 'ERECTED', 'PURPOSE', 'LENGTH', 'LANESCLEAR', 'T-OR-D', 'MATERIAL', 'SPAN', 'REL-L', 'TYPE'])
data['LENGTH'] = data['LENGTH'].apply(pd.to_numeric, errors='coerce')
data = data.groupby(['PURPOSE', 'MATERIAL']).agg({'LENGTH': "mean"}).rename(columns={'LENGTH': 'Average length'})
data
```

```python
# Reading the data set
# Convert LENGTH column to numeric if failed return NaN
# Group by ERECTED with given range of values and use mean, count, max and min agg functions on LENGTH column then rename it after flatten the column levels. 
data = pd.read_csv(CSV_file_source, index_col=False, header=None, names=['IDENTIF', 'RIVER', 'LOCATION', 'ERECTED', 'PURPOSE', 'LENGTH', 'LANESCLEAR', 'T-OR-D', 'MATERIAL', 'SPAN', 'REL-L', 'TYPE'])
data['LENGTH'] = data['LENGTH'].apply(pd.to_numeric, errors='coerce')
data = data.groupby(pd.cut(data["ERECTED"], np.arange(1818, 2028, 42))).agg(
    {
        'LENGTH': ["mean", "count", "max", "min"]
    })
data.columns = data.columns.get_level_values(1)
data = data.rename(columns={'mean': "Average Length", 'count': 'Count', 'max': 'Max Length', 'min': 'Min Length'})
data
```
