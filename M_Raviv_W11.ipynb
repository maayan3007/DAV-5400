```python
import requests
import json
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

API_KEY = "K7y2YIKVycSKvmPtAKtxDQEN8cWLk0n7"

# Let's access most viewed articles on NYTimes.com API - endpoint: https://developer.nytimes.com/docs/most-popular-product/1/routes/viewed/%7Bperiod%7D.json/get
period = int(input("Enter period in days(1, 7 or 30 only): ").strip()) # Get user input for only parameter for the endpoint
```
