## Introduction
<br>
The Affordable Care Act (ACA) is the name for the comprehensive health care reform law and its amendments which addresses health insurance coverage, health care costs, and preventive care. The law was enacted in two parts: The Patient Protection and Affordable Care Act was signed into law on March 23, 2010, by President Barack Obama and was amended by the Health Care and Education Reconciliation Act on March 30, 2010.
The Affordable Care Act (ACA) led to historic gains in health insurance coverage by extending Medicaid coverage to many low-income individuals and providing Marketplace subsidies for individuals below 400% of poverty. The number of uninsured nonelderly Americans decreased from over 44 million in 2013 (the year before the major coverage provisions went into effect) to just below 27 million in 2016. 
Many states have chosen not to expand Medicaid since the US Federal government will not be covering the entire cost of that expansion. During the first couple years of Medicaid expansion, the federal government paid 100 percent of the costs. That funding started winding down in 2016 and will drop to 90 percent in 2020, meaning, the other 10% must be covered by the states themselves. As such, states that have opted to expand Medicaid coverage are also having to find new sources of revenue to pay for this expansion. 
The subsidies also rely on payments of penalties by people who decide not to purchase any type of health insurance. One unfortunate result of the ACA is that insurance premiums and deductibles have skyrocketed for everyone who is ineligible for the subsidies. This has led many relatively healthy people who don't qualify for the subsidies to entirely drop their health insurance since it is much less expensive to simply pay the Federal penalty for not having insurance than it is to pay the very expensive insurance premiums + deductibles that resulted from the ACA. 
The purpose of this research is to indicate whether the Obama care plan succeeds its goal to make coverage more affordable to more people. These subsidies are paid for by Obamacare taxes on higher income families and some health care providers. It also required everyone to get insurance.
We will review the uninsured rate for each state according to the population size including other factors as Medicare and Medicaid plan enrollment which are the official plans of the ACA.

__USA Choropleth Map - Uninsured Rate Changes by State__

In the choropleth map below, you can see that the West Coast region has the highest change of uninsured rate, while the Mid-West region has the lowest change of uninsured rate. 

```python
import plotly
import plotly.graph_objs as go

plotly.offline.init_notebook_mode(connected=True)

ChoroplethMapData = dataset.copy() # take a copy of dataset
# Add text coloum to indicate the text which should show on each state while hovering
ChoroplethMapData['text'] = ChoroplethMapData['State'] + " <br>Change:  " + ChoroplethMapData['Uninsured Rate Change 2010-2015'].astype(str) + "%"

scl = [
    [0.0, 'rgb(242,240,247)'],
    [0.2, 'rgb(218,218,235)'],
    [0.4, 'rgb(188,189,220)'],
    [0.6, 'rgb(158,154,200)'],
    [0.8, 'rgb(117,107,177)'],
    [1.0, 'rgb(84,39,143)']
]

data = [
    go.Choropleth(
        colorscale = scl,
        autocolorscale = False,
        locations = ChoroplethMapData['Abbr'],
        z = ChoroplethMapData['Uninsured Rate Change 2010-2015'],
        locationmode = 'USA-states',
        text = ChoroplethMapData['text'],
        marker = go.choropleth.Marker(
            line = go.choropleth.marker.Line(
                color = 'rgb(255,255,255)',
                width = 2
            )),
        colorbar = go.choropleth.ColorBar(
            title = "Rate Change as percentage")
    )
]

layout = go.Layout(
    title = go.layout.Title(
        text = 'USA Uninsured Rate Change by State between 2010-2015'
    ),
    geo = go.layout.Geo(
        scope = 'usa',
        projection = go.layout.geo.Projection(type = 'albers usa'),
        showlakes = True,
        lakecolor = 'rgb(255, 255, 255)'),
)

fig = go.Figure(data = data, layout = layout)

plotly.offline.iplot(fig)
```

### 4. How has the Affordable Care Act changed the rate of citizens with health insurance coverage? 
For this research question, we calculated the actual number of uninsured people in each year according to the population data in each state and the rate of uninsured. 


```python
dataset["Uninsured_2010"] = dataset["Uninsured Rate 2010"]*dataset["Population_2010"]/100
dataset["Uninsured_2010"] = dataset["Uninsured_2010"].apply(lambda x: int(x))
dataset["Uninsured_2015"] = dataset["Uninsured Rate 2015"]*dataset["Population_2015"]/100
dataset["Uninsured_2015"] = dataset["Uninsured_2015"].apply(lambda x: int(x))
dataset["UninsuredChange_2010_2015"] = dataset["Uninsured_2015"]-dataset["Uninsured_2010"]
dataset.head()
```

```python
#this a function that calculate the mean, sum and median for the uninsured variables. 
pop_data = dataset[['State Medicaid Expansion 2013', 'Population_2015', 'Population_2010', 'Uninsured_2015', 'Uninsured_2010']]
pop_data = pop_data.groupby(['State Medicaid Expansion 2013'], as_index=False).agg({
    
    'Uninsured_2015': ['sum', 'median', 'mean'],
    'Uninsured_2010': ['sum', 'median', 'mean'],
})
pop_data
```

### Compare  Number of Uninsured - Median
According to the new variables that were created we can see below that there was a decrease in the median the number of people who were not insured in the year 2015 compared to 2010. 
<br>
Among the group of states which accept the expansion, the number of uninsured was lower compared to the states who didn't accept the expansion,  273K comparing to 490K.

```python
# Compare Uninsured Rate - median
pop_2010 = pop_data["Uninsured_2010"]['median']
pop_2015 = pop_data["Uninsured_2015"]['median']

ind = np.arange(len(pop_2010))  # the x locations for the groups
width = 0.35  # the width of the bars

def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0.5, 'right': 0.1, 'left': .9}  # x_txt = x + w*off

    for rect in rects:
        height = round(rect.get_height(), 2)
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 1.01*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')
        
fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, pop_2010, width, color='SkyBlue', label='Uninsured_2010')
rects2 = ax.bar(ind + width/2, pop_2015, width, color='IndianRed', label='Uninsured_2015')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Uninsured in population')
ax.set_xlabel('Expansion')
ax.set_title('Uninsured Rate 2010-2015 Median | expansion True-False')
ax.set_xticks(ind)
ax.set_xticklabels(('False', 'True'))
ax.legend()
ax.margins(x=.1, y=.2)
autolabel(rects1, "left")
autolabel(rects2, "right")
```

### Compare  Number of Uninsured - Mean
According to the new variables that were created we can see below that there was a decrease in the median the number of people who were not insured in the year 2015 compared to 2010. 
<br>
Among the group of states which accept the expansion, the number of uninsured was lower compared to the states who didn't accept the expansion,  462K comparing to 812K. 

```python
# Compare Uninsured Rate - mean
pop_2010 = pop_data["Uninsured_2010"]['mean']
pop_2015 = pop_data["Uninsured_2015"]['mean']

fig, ax = plt.subplots()
rects3 = ax.bar(ind - width/2, pop_2010, width, color='SkyBlue', label='Uninsured_2010')
rects4 = ax.bar(ind + width/2, pop_2015, width, color='IndianRed', label='Uninsured_2015')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Uninsured in population')
ax.set_xlabel('Expansion')
ax.set_title('Uninsured Rate 2010-2015 mean | expansion True-False')
ax.set_xticks(ind)
ax.set_xticklabels(('False', 'True'))
ax.legend()
ax.margins(x=.1, y=.4)
autolabel(rects3, "left")
autolabel(rects4, "right")


plt.show()
```
